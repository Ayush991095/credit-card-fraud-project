```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/creditcard.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
fraud = df[df['Class'] == 1]
non_fraud = df[df['Class'] == 0]
percentage_of_fraud = len(fraud)/(len(df))*100
print(f'percentage_of_fraud = {percentage_of_fraud}%')
print(f'fraud = {len(fraud)}')
print(f'non_fraud = {len(non_fraud)}')
```

```python
labels = ['Non Fraud', 'Fraud']
sizes = [len(non_fraud), len(fraud)]
plt.bar(labels, sizes)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
df['scaled_amount'] = sc.fit_transform(df['Amount'].values.reshape(-1,1))
```

```python
df
```

```python
df.drop(['Amount'], axis = 1, inplace = True)
```

```python
df.drop(['Time'], axis = 1, inplace = True)
```

```python
df
```

```python
x = df.drop(['Class'], axis = 1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
from sklearn.metrics import accuracy_score
acc_sco = accuracy_score(y_test, y_predict)*100
print(f'accuracy_score_DT = {acc_sco}%')
```

```python
from sklearn.metrics import *
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True, cmap = 'Blues', fmt = 'd' )
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
acc_sco = accuracy_score(y_test, y_predict)*100
print(f'accuracy_score_RF = {acc_sco}%')
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True, cmap = 'Greens', fmt = 'd')
```
